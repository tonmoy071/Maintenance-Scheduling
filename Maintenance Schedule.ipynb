```python
machine_ids = machine_ids_str.split(',')
categories = categories_str.split(',')
last_maintenance_dates = last_maintenance_str.split(',')
machines = [
    {
        "ID": machine_id, 
        "Category": category, 
        "LastMaintenance": datetime.strptime(last_maintenance, "%d-%b-%y")
    }
    for machine_id, category, last_maintenance in zip(machine_ids, categories, last_maintenance_dates)
]
```

```python
start_date = datetime(2024, 5, 1)
end_date = datetime(2024, 6, 30)
def get_valid_dates(start_date, end_date):
    valid_dates = []
    current_date = start_date
    while current_date <= end_date:
        if current_date.weekday() != 4:
            valid_dates.append(current_date)
        current_date += timedelta(days=1)
    return valid_dates
valid_dates = get_valid_dates(start_date, end_date)
intervals = {
    "C": (7, 9),
    "B": (11, 13),
    "A": (32, 34)
}
def calculate_average_maintenance_time(dates):
    if len(dates) < 2:
        return np.nan
    intervals = [(dates[i] - dates[i - 1]).days for i in range(1, len(dates))]
    return np.mean(intervals)
```

```python
schedule_list = []
def calculate_average_maintenance_time(dates):
    if len(dates) < 2:
        return np.nan
    intervals = [(dates[i] - dates[i - 1]).days for i in range(1, len(dates))]
    return np.mean(intervals)
for machine in machines:
    category = machine["Category"]
    last_maintenance = machine["LastMaintenance"]
    lower_bound, upper_bound = intervals.get(category, (7, 9))
    maintenance_dates = [last_maintenance]
    while True:
        next_maintenance = maintenance_dates[-1] + timedelta(days=np.random.randint(lower_bound, upper_bound + 1))
        if next_maintenance > end_date:
            break
        if next_maintenance in valid_dates:
            maintenance_dates.append(next_maintenance)
    for i, date in enumerate(maintenance_dates):
        schedule_list.append({
            "Machine ID": machine["ID"],
            "Category": category,
            "Maintenance Date": date.strftime("%m/%d/%Y") + f" ({date.strftime('%A')})",
            "Average_Maintenance_Time": calculate_average_maintenance_time(maintenance_dates) if i > 0 else np.nan,
            "Latest_Maintenance_Date": date.strftime("%m/%d/%Y")
        })
```

```python
df_schedule = pd.DataFrame(schedule_list)
schedule_grouped = df_schedule.groupby("Maintenance Date").size()
excess_dates = schedule_grouped[schedule_grouped > 16].index
for excess_date in excess_dates:
    excess_machines = df_schedule[df_schedule["Maintenance Date"] == excess_date]
    if len(excess_machines) > 16:
        indices_to_remove = excess_machines.index[16:]
        df_schedule = df_schedule.drop(indices_to_remove)
```

```python
maintenance_schedule = df_schedule.groupby(["Machine ID", "Category"]).agg(
    Maintenance_Dates=("Maintenance Date", lambda x: ', '.join(x)),
    Average_Maintenance_Time=("Average_Maintenance_Time", 'mean')
).reset_index()
latest_dates = df_schedule.groupby("Machine ID").agg(
    Latest_Maintenance_Date=("Maintenance Date", lambda x: max(pd.to_datetime(x, format='%m/%d/%Y (%A)')).strftime('%m/%d/%Y'))
).reset_index()
maintenance_schedule = maintenance_schedule.merge(latest_dates, on="Machine ID")
maintenance_schedule = maintenance_schedule.sort_values(by="Category")
maintenance_schedule.to_csv("machine_maintenance_schedule.csv", index=False)
summary = df_schedule.groupby("Maintenance Date").agg(
    Total_Preventive_Maintenance=("Machine ID", "size"),
    Machine_IDs=("Machine ID", lambda x: ', '.join(x))
).reset_index()
summary.rename(columns={"Maintenance Date": "Date"}, inplace=True)
summary = summary.sort_values(by="Date")
summary.to_csv("maintenance_summary.csv", index=False)
print("Maintenance Schedule:")
print(maintenance_schedule)
print("\nMaintenance Summary:")
print(summary)
```
